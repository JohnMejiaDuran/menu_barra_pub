```python
import random
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import Audio

imagen_ahorcado = cv2.imread('ahorcado.jpg',1)
nuevo_ancho = 250
nuevo_alto =250
tamaño_imagen = cv2.resize(imagen_ahorcado,(nuevo_ancho,nuevo_alto))
def obtener_palabra_aleatoria():
    palabras = ["python", "programacion", "juego", "ahorcado", "computadora", "aprender","perro","pato"]
    return random.choice(palabras)



def jugar_ahorcado():
    palabra = obtener_palabra_aleatoria()
    letras_adivinadas = []
    intentos = 6
    longitud=len(palabra)
    print(f"la palabra tiene {longitud} letras")
    while True:
        letra = input("Ingresa una letra: ").lower()



        if letra in letras_adivinadas:
            print("Ya has adivinado esa letra. Intenta nuevamente.")
            continue

        if letra in palabra:
            letras_adivinadas.append(letra)
            palabra_oculta = " "

            for char in palabra:
                if char in letras_adivinadas:
                    palabra_oculta += char
                else:
                    palabra_oculta += "_"

            print("¡Correcto! Letras adivinadas:", " ".join(letras_adivinadas))
            print("Palabra oculta:", palabra_oculta)

            if "_" not in palabra_oculta:
                print("¡Felicidades! Has adivinado la palabra correctamente.")
                tts = gTTS("Ganaste",lang="es")
                tts.save('1.wav')
                sound_file = '1.wav'
                display(Audio(sound_file, autoplay=True))
                break
        else:
            intentos -= 1
            print("Letra incorrecta. Te quedan {} intentos.".format(intentos))

            if intentos == 0:
                print("¡Oh no! Te has quedado sin intentos. La palabra era:", palabra)

        if intentos == 6:
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')
        if intentos == 5:
          cv2.circle(tamaño_imagen,(170,135),5,(255,0,0),(20))
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')
        if intentos == 4:
          cv2.line(tamaño_imagen,(170,140),(170,175),(255,0,0),(3))
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')
        if intentos == 3:
          cv2.line(tamaño_imagen,(170,140),(150,150),(255,0,0),(3))
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')
        if intentos == 2:
          cv2.line(tamaño_imagen,(170,140),(190,150),(255,0,0),(3))
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')
        if intentos == 1:
          cv2.line(tamaño_imagen,(170,175),(150,195),(255,0,0),(3))
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')
        if intentos == 0:
          cv2.line(tamaño_imagen,(170,175),(190,195),(255,0,0),(3))
          cv2_imshow(tamaño_imagen)
          print(f'Te quedan {intentos} intentos')

print("¡Bienvenido al juego del ahorcado!")
jugar_ahorcado()
```
